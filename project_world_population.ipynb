```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. 다운로드한 CSV 파일을 읽어 DataFrame pop_df로 저장하고, 처음 5행을 출력하시오.
pop_df = pd.read_csv('world_population.csv')
pop_df.head()
```

```python
# 2. 데이터셋의 정보(info() 이용)와 요약통계(describe() 이용) 결과를 출력하시오.
pop_df.info()
pop_df.describe()
```

```python
# 3. 결측값이 있는 열(column)과, 각 결측값 개수를 출력하시오.
print(pop_df.isna().sum())
print(pop_df.columns[pop_df.isna().any(axis=0)])
```

```python
# 4. 각 열에 대한 자료형(dtypes)을 확인하시오.
pop_df.dtypes
```

```python
# 5. 대륙별(Continent) 국가의 개수를 구하시오.
pop_df["Continent"].value_counts()
```

```python
# 6. 국가 이름을 기준으로 알파벳 오름차순으로 정렬하시오.
pop_df.sort_values(by="Country/Territory")
```

```python
# 7. 2022년 인구(2022 Population)를 기준으로 상위 10개국을 출력하시오.
pop_df.sort_values(by="2022 Population", ascending=False).head(10)
```

```python
# 8. 인구 증가율(Growth Rate)이 1% 이상인 국가를 출력하시오.
pop_df[pop_df["Growth Rate"] >= 1]
```

```python
# 9. 인구 밀도(Density (per km2))가 1000 이상인 국가를 출력하시오.
pop_df[pop_df["Density (per km²)"] >= 1000]
```

```python
# 10. 면적이 1,000,000km2 이상이고 인구 1악 이상인 나라의 이름과 면적, 인구를 출력하시오.
cond = (pop_df["Area (km²)"] >= 1_000_000) & (pop_df["2022 Population"] >= 100_000_000)
pop_df.loc[cond, ["Country/Territory", "Area (km²)", "2022 Population"]]
```

```python
# 11. ‘Asia’ 대륙 국가들의 평균 인구를 구하시오.
asia_mean_pop = pop_df.loc[pop_df["Continent"] == "Asia", "2022 Population"].mean()
print(asia_mean_pop)
```

```python
# 12. 2022년 인구에서 2021년 인구를 뺀 “Population Change”라는 이름의 새로운 열을 만드시오.
pop_df["Population Change"] = pop_df["2022 Population"] - pop_df["2020 Population"]
pop_df["Population Change"]
```

```python
# 13. 위에서 생성한 열을 기준으로 인구 증가량이 가장 많은 상위 5개국을 출력하시오.
pop_df.sort_values(by="Population Change", ascending=False).head(5)[["Country/Territory", "Population Change"]]
```

```python
# 14. 대륙별 평균 인구 증가율을 구하시오.
pop_df.groupby("Continent")["Growth Rate"].mean()
```

```python
# 15. 대륙별 전체인구(2022년 Population) 합계를 구하고 내림차순으로 출력하시오.
continent_pop_sum = pop_df.groupby("Continent")["2022 Population"].sum().sort_values(ascending=False)
continent_pop_sum
```

```python
# 16. 인구 밀도가 가장 높은 10개국을 출력하시오.
pop_df.sort_values(by="Density (per km²)", ascending=False).head(10)[["Country/Territory", "Density (per km²)"]]
```

```python
# 17. 인구 대비 면적비율(=면적/인구)이 가장 작은 5개국을 출력하시오.
pop_df["Area_per_Pop"] = pop_df["Area (km²)"] / pop_df["2022 Population"]

pop_df.sort_values(by="Area_per_Pop").head(5)[["Country/Territory", "Area (km²)", "2022 Population", "Area_per_Pop"]]
```

```python
# 18. 대륙명을 약어로 바꾸어 pop1_df로 저장하시오. 예: Asia  AS, Europe  EU, Africa  AF 등
cont_map = {"Asia": "AS", "Europe": "EU", "Africa": "AF", "North America": "NA", "South America": "SA", "Oceania": "OC", "Antarctica": "AN"}
pop1_df = pop_df.copy()
pop1_df["Continent"] = pop1_df["Continent"].replace(cont_map)
pop1_df.head()
```

```python
# 19. 2022년 인구에 대한 기술 통계(describe())를 구하시오.
pop_df["2022 Population"].describe()
```

```python
# 20. 인구의 25%, 50%, 75% 분위값을 구하시오.
pop_df["2022 Population"].quantile([0.25, 0.5, 0.75])
```

```python
# 21. 인구(2022 Population)와 면적(Area (lkm2)), 인구 밀도(Density (km2))간의 상관 계수를 구하시오.
pop_df[["2022 Population", "Area (km²)", "Density (per km²)"]].corr()
```

```python
# 22. 다음의 링크를 통해 2022년 전세계 국가별 GDP 데이터 파일을 다운로드하고 읽어 들여 dgp_df DataFrame을 만드시오.
dgp_df = pd.read_csv('GDP.csv')
dgp_df.head()
```

```python
# 23. 위의 dgp_df과 pop_df를 국가이름(Country)을 기준으로 병합한 popdgp_df를 생성한 후, 처음 5행을 출력하시오.
popdgp_df = pd.merge(pop_df, dgp_df, left_on="Country/Territory", right_on="Country", how="inner")
popdgp_df.head()
```

```python
# 24. GDP 상위 10개국의 GDP 대비 인구비율(GDP_per_capita)을 출력하시오.
popdgp_df["GDP_2022"] = (popdgp_df["GDP(nominal,2022)"].str.replace('[$,]', '', regex=True).astype(float))
popdgp_df["GDP_per_capita"] = popdgp_df["GDP_2022"] / popdgp_df["2022 Population"]
gdp_top10 = popdgp_df.sort_values(by="GDP_2022", ascending=False).head(10)
gdp_top10[["Country/Territory", "GDP_per_capita"]]
```
